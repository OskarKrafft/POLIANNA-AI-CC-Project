```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = { 'queries' : ["Renewables", "Climate", "Hydro","Wind","Solar", "H-Fuel Cells", "Distributed", "Batteries", "EV", "Biothermal", "Heating/Cooling" ],
'2000-2019' : [230,247,24,34,28,46,74,193,101,91,29],
'2020-2024' : [96, 135, 9, 17, 16, 30, 27, 71, 37, 34, 17]}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Avg_Documents_2000_19'] = df['2000-2019'] / 20
df['Avg_Documents_2020_24'] = df['2020-2024'] / 5
df['Avg_Increase'] = round(df['Avg_Documents_2020_24'] / df['Avg_Documents_2000_19'] * 100)

df
```

```python
# Create a figure and two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot the first pie chart
ax1.pie(df['2000-2019'], labels=df['queries'], autopct='%1.1f%%')
ax1.set_title('Document Topic Breakdowns 2000 - 2019')

# Plot the second pie chart
ax2.pie(df['2020-2024'], labels=df['queries'], autopct='%1.1f%%')
ax2.set_title('Document Topic Breakdowns 2020 - 2024')

# Show the plot
plt.show()
```
